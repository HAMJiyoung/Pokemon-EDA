# * 가장 강한 포켓몬 1가지는?
### 한정된 자원으로 1가지 포켓몬만 키울 수 있다면, 어떤 것을 키우는게 가장 많은 포켓몬을 상대로 이길 수 있을까?

포켓몬을 기반으로한 게임은 꽤 많아서, 계산식과 기본 능력치 기준이 다 다르다. 너무 복잡해지는 것을 막기 위해 몇가지를 통일하였다.

1. 능력치 
 - 레벨이 올라가면 더 좋은 능력치를 갖게 되나, 여기서는 **레벨을 고려하지 않은 기본 능력치를 기준**으로 계산한다. 
 

2. 진화형태 
 - 이상해씨->이상해풀->이상해꽃 등과 같은 진화 형태는 각자의 개체로 본다.
 
 
3. 1세대를 기준으로 
 
- 데미지 = (위력 × 공격 × (레벨 × [[급소]] × 2 ÷ 5 + 2 ) ÷ 방어 ÷ 50 + 2 ) × [[자속 보정]] × 타입상성1 × 타입상성2 × 랜덤수/255

- 위력 = [[도우미]] × 기술위력 × 도구보정 × [[충전]] × [[흙놀이]] × [[물놀이]] × 특성보정 × 상대특성보정
  - 도우미, 도구보정, 충전, 흙놀이, 물놀이, 특성보정, 상태특성보정은 없음. 따라서 위력 = 기술위력인데, 기술의 종류는 745개. 포켓몬의 종류와 기술의 종류의 경우의 수를 따져보면 너무 복잡해져서 해당 변수 삭제. 
  
- 공격 = 스탯 × [[랭크 (상태변화)|랭크]]보정 × [[특성]]보정 × [[도구]]보정
  - 랭크(기술, 도구에 따른 버프 효과), 특성, 도구 없음. 따라서 공격 = 스탯 
  - 특수 공격은 공격 3번 사용 시 1번 사용할 수 있도록 제한. 특수공격이니까. 

- 레벨 
  - 1로 통일 
 
- 급소 
  - 아래와 같이 계산되나, 1세대 제외 모두 6.25%가 기본으로 6.25%로 통일  
  - 급소율의 계산
    > 6세대
기본 = 1/16 (6.25%)
1랭크 = 1/8 (12.5%)
2랭크 = 1/2 (50%)
3랭크 = 1/1 (100%)
2~5세대
기본 = 1/16 (6.25%)
1랭크 = 1/8 (12.5%)
2랭크 = 1/4 (25%)
3랭크 = 1/3 (33.3%)
4랭크 = 1/2 (50%)
1세대
일반 기술 = 스피드/512
급소 보정 기술 = 스피드/64

- 특수방어 = 스탯 × [[랭크]]보정 × Mod × 자폭보정
  - 랭크(기술, 도구에 따른 버프 효과), Mod(특성, 상태, 기술, 날씨 같은 것), 자폭보정 없음. 따라서 방어 및 특수방어 = 스탯 
  - 특수 방어는 특수 공격 시 발동되는 방어력. 
 
- 자속보정
  - 사용하는 기술의 타입과 사용하는 포켓몬의 타입이 일치하면 1.5를 대입. 사용하는 기술의 타입과 사용하는 포켓몬의 타입이 일치하면서 적응력 특성의 포켓몬이라면 2를 대입.
  - 기술은 포켓몬에 따라 랜덤으로 나오고, 배우기 등을 고려하면 복잡하기 때문에 기술 타입과 포켓몬 타입이 일치하고 적응력 등의 특성은 고려하지 않는 것을 전제로 **1.5로 통일.** 
 

- 타입상성1, 타입상성2
  - https://pokemon.fandom.com/ko/wiki/%ED%83%80%EC%9E%85_%EC%83%81%EC%84%B1%ED%91%9C
  - 상성은 공격하는 기술 : 방어하는 포켓몬의 타입 기준이다.
  - 타입이 1개인 포켓몬의 경우는 타입 상성표에 있는 그대로 계산하면 된다. 
  - 타입이 2개인 포켓몬은 각 타입의 해당하는 배율을 서로 곱한 결과를 사용한다. 
  - 여기서는 **타입이 1개인 포켓몬은 타입상성표 그대로 계산**하고, **타입이 2개인 포켓몬들은 상대방에게 효과가 더 큰 타입으로 골라서 사용**하는 것으로 한다.
  > 예를 들어 불꽃/비행타입인 리자몽에게 바위타입의 기술을 사용하면 불꽃타입과 비행타입 각각 2배의 데미지를 받기 때문에 총 4배의 데미지를 받게 된다. 
반대로 리자몽에게 땅타입의 기술을 사용하면 불꽃타입은 2배, 비행타입은 0배의 데미지를 받기 떄문에 0배의 데미지, 즉 아무런 데미지를 받지 않는다.
따라서 불꽃/비행타입인 리자몽과 땅/바위 타입 딱구리가 싸운다면 딱구리는 땅 타입 기술 사용 시 0배, 바위 타입 기술 사용 시 4배의 데미지를 줄 수 있기 때문에 바위 타입 기술로 공격할 것이다.
